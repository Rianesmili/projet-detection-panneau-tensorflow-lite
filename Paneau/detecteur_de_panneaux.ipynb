On telecharge et dezip le fichier qui contient notre dataset

```python
#Import dataset
import pathlib
import os
data_dir = tf.keras.utils.get_file(
    "dataset.zip",
    "https://github.com/Rianesmili/projet-detection-panneau-tf/blob/main/Paneau/dataset.zip?raw=true",
    extract=False)

import zipfile
with zipfile.ZipFile(data_dir, 'r') as zip_ref:
    zip_ref.extractall('/content/datasets')

data_dir = pathlib.Path('/content/datasets/dataset')
print(data_dir)
print(os.path.abspath(data_dir))
```

```python
#on liste le nombre d'l'ement de notre dataset
image_count = len(list(data_dir.glob('*/*')))
print(image_count)
```

```python
batch_size = 3
img_height = 200
img_width = 200
```

Dans ce code il ne permet que chaque classe correspond a un folder du meme nom

Construire le jeux entrainement


```python
train_data = tf.keras.preprocessing.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="training",
  seed=42,
  image_size=(img_height, img_width),
  batch_size=batch_size,
  )

```

Construire le jeux de validation




```python
val_data = tf.keras.preprocessing.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="validation",
  seed=42,
  image_size=(img_height, img_width),
  batch_size=batch_size)

class_names = val_data.class_names
print(class_names)
```

Un echantillion correspandant au batch size on a batch size = 4 on vas avoir quatres images

```python
plt.figure(figsize=(10, 10))
for images, labels in train_data.take(1):
  for i in range(3):
    ax = plt.subplot(1, 4, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[labels[i]])
    plt.axis("off")
```

Dfinition du reseau de neurones convolutionnel



```python
from tensorflow.keras import layers

num_classes =  3 # danger, indication, obligation d'ou le 3

model = tf.keras.Sequential([ #kerasSequantial qui vas nous servir de conteneur
    layers.experimental.preprocessing.Rescaling(1./255), #rescaling
                              #definir les differantes couches
    layers.Conv2D(128,4, activation='relu'),  #couche de convolution
    layers.MaxPooling2D(), # couche de pooling
    layers.Conv2D(64,4, activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32,4, activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(16,4, activation='relu'),
    layers.MaxPooling2D(), #max pooling prend le maximum a chaque fois il exsite d'autres qui prennent la moyenne avrege
    #ces couches seront une suite de convolution et de poolig aui seront de plus en plus petit
    layers.Flatten(), #flatten permet de construire un vecteur a partir de la matrice
    layers.Dense(64,activation='relu'),
    layers.Dense(num_classes, activation='softmax') #softmax pour avoir la probabilit entre les differantes classes
])

```

On vas alors compiler notre model que l'on a dfini

```python
model.compile(optimizer='adam',
              loss=tf.losses.SparseCategoricalCrossentropy(from_logits=True),
  metrics=['accuracy'],)
```

```python
logdir="logs"

tensorboard_callback = keras.callbacks.TensorBoard(log_dir=logdir,histogram_freq=1, write_images=logdir,
                                                   embeddings_data=train_data)
```

```python
model.fit(
    train_data,
  validation_data=val_data,
  epochs=20, #on entraine le model sur 20 epochs
  callbacks=[tensorboard_callback]
)
```

```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt

file_to_predict = files.upload()

for file_ in file_to_predict:
    image_to_predict = cv2.imread(file_, cv2.IMREAD_COLOR)
    plt.imshow(cv2.cvtColor(image_to_predict, cv2.COLOR_BGR2RGB))
    plt.show()

    img_to_predict = np.expand_dims(cv2.resize(image_to_predict, (200, 200)), axis=0)
    predictions = model.predict(img_to_predict)
    class_index = np.argmax(predictions)  # Obtenez l'index de la classe prdite

    if class_index == 0:
        # Si la classe prdite est "panneau de limitation"
        plt.imshow(cv2.cvtColor(image_array_limitation, cv2.COLOR_BGR2RGB))
        plt.show()
        print("C'est un panneau de limitation !")
    elif class_index == 1:
        # Si la classe prdite est "panneau obligation"
        plt.imshow(cv2.cvtColor(image_obligation, cv2.COLOR_BGR2RGB))
        plt.show()
        print("C'est un panneau d'obligation !")
    elif class_index == 2:
        # Si la classe prdite est "panneau sens interdit"
        plt.imshow(cv2.cvtColor(image_sens_interdit, cv2.COLOR_BGR2RGB))
        plt.show()
        print("C'est un panneau indiquant un sens interdit !")
    elif class_index == 3:
        # Si la classe prdite est "panneau stationnement interdit"
        plt.imshow(cv2.cvtColor(image_stationnement_interdit, cv2.COLOR_BGR2RGB))
        plt.show()
        print("C'est un panneau de stationement interdit !")
```

```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Fonction de prtraitement de l'image
def preprocess_image(file):
    image = cv2.imread(file, cv2.IMREAD_COLOR)
    image = cv2.resize(image, (200, 200))  # Redimensionner l'image  la taille attendue
    image = image / 255.0  # Normalisation des valeurs de pixel
    return np.expand_dims(image, axis=0)  # Ajouter une dimension pour correspondre  l'entre du modle

# Charger l'image uploade
uploaded = files.upload()

for file_name in uploaded.keys():
    file_path = file_name
    image = preprocess_image(file_path)

    # Prdiction avec le modle
    predictions = model.predict(image)
    predicted_class = np.argmax(predictions[0])

    # Afficher l'image avec la prdiction
    plt.imshow(cv2.cvtColor(cv2.imread(file_path), cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title(f'Predicted Class: {predicted_class}')
    plt.show()
```
